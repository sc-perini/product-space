```python
# descargo lista completa de paÃ­ses
country_list = pd.read_csv(r'UN Comtrade Country List.csv')
country_list
```

```python
# armo lista de paÃ­ses para hacer la consulta en Comtrade
len(country_list.ctyCode)
countries = country_list.ctyCode
print(type(countries.to_list()))
countries_string = [str(x) for x in  countries.to_list()]
print(countries_string)
```

```python
# descargo lista completa de bienes
lista_bienes = pd.read_excel(r'UN Comtrade Commodity Classifications.xlsx')
lista_bienes
```

```python
# selecciono codigo de bienes a 4 dÃ­gitos del sistema armonizado 
# H1 corresponde a la nomenclatura de 1996 para poder comparar con aÃ±os posteriores
# idea comparar un promedio de aÃ±os para los Ãºltimos 20 aÃ±os: 1997-99, 2007-09 y 2017-19 
lista_bienes[lista_bienes['Code'].map(len) == 4].Classification.unique()
```

```python
# selecciono 4 digitos y H1
bienes_4dig = lista_bienes[(lista_bienes['Code'].map(len) == 4) & (lista_bienes['Classification'] == 'H1')]
bienes_4dig
```

```python
# armo lista para hacer consulta en comtrade
cod_bs_4dig = bienes_4dig.Code.to_list()
len(cod_bs_4dig)
```

```python
type(cod_bs_4dig)
cod_bs_4dig[1239:]
```

```python
# pruebo con una consulta para todos los productos con un paÃ­s y un aÃ±o
comapi.getComtradeData(comcodes=cod_bs_4dig[1239:], reporter=['32'], partner=['0'], years=['2019'], freq='A', rg=['1'], fmt='json', rowmax=50000)
```

```python
# Armo consulta para iterar sobre 3 aÃ±os
# consulto todos los bienes a 4 dÃ­gitos (H1)
# reporter = todos los paÃ­ses
# partner = Mundo
# trade flow = imports 
year = ['2017','2018','2019']
appended_data = []
for y in range(0,3):
        data_prod = comapi.getComtradeData(comcodes=cod_bs_4dig, reporter=countries_string, partner=['0'], years=[year[y]], freq='A', rg=['1'], fmt='json', rowmax=50000)
        # store DataFrame in list
        appended_data.append(data_prod)
        # see pd.concat documentation for more info
appended_data = pd.concat(appended_data)
```
